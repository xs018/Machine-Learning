```python
import  matplotlib.pyplot  as  plt 
import  numpy  as  np 
import  pandas  as  pd 
from  matplotlib.pyplot import  figure
```

```python
#t1 is the total training time of last 9 epochs by 1 GPU
#t2 is the total training time of last 9 epochs by 2 GPU
#t4 is the total training time of last 9 epochs by 3 GPU
#The first location is batch size 256, then 512, last location is 1024
t1  =  [26,  26,  59] 
t2  =  [42,  33,  50] 
t4  =  [21,  43,  72] 
t8  =  [52,  21,  29]
```

```python
plotdata  =  pd.DataFrame({ "Batch Size = 256":[t1[0],  t2[0],  t4[0],  t8[0]],
                           "Batch Size = 512":[t1[1],  t2[1],  t4[1],  t8[1]],
                           "Batch Size = 1024":[t1[2],  t2[2],  t4[2],  t8[2]]}, 
                          index=["1 GPU",  "2 GPUs",  "4 GPUs",  "8 GPUs"]) 
plotdata.plot(kind="bar",figsize=(15,  8)) 
plt.title("Multi GPUs Performance") 
plt.xlabel("Number of GPUs") 
plt.ylabel("Computational Time (s/epoch)") 
plt.legend(loc='best')
```

```python
speedup_256  =  [t1[0]/t1[0],  t1[0]/t2[0], t1[0]/t4[0], t1[0]/t8[0]] 
speedup_512  =  [t1[1]/t1[1],  t1[1]/t2[1], t1[1]/t4[1], t1[0]/t8[0]] 
speedup_1024  =  [t1[2]/t1[2],  t1[2]/t2[2], t1[2]/t4[2], t1[0]/t8[0]] 
GPU_numb  =  [1,2,4,8] 
```

```python
figure(figsize=(8,  6),  dpi=80) 
plt.plot(GPU_numb,  speedup_256,  markersize=8, marker='o',  markerfacecolor='blue',   color='skyblue',  linewidth=2,  linestyle='dashed',  label="Batch Size = 256") 
plt.plot(GPU_numb,  speedup_512,  markersize=8, marker='D',  markerfacecolor='orange', color='orange',  linewidth=2,  linestyle='dashed',  label="Batch Size = 512") 
plt.plot(GPU_numb,  speedup_1024,  markersize=8, marker='*',  markerfacecolor='green', color='green',  linewidth=2,  linestyle='dashed',  label="Batch Size = 1024") 


plt.xlabel("Number of GPUs") 
plt.ylabel("Speedup") 
plt.title("Speedup on multiple GPUs of Bridges2 GPU-AI") 
```
